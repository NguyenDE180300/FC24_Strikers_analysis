## Import lib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

## Load file csv

```python
df = pd.read_csv("D:\Workplace\.FALL2024\MAS\Final_Prj\st_players.csv")
```

```python
df.info()
```

```python
left_footed = df['foot_Left'].sum()
right_footed = df['foot_Right'].sum()

plt.figure(figsize= (10, 8))
plt.bar(['Left Footed', 'Right Footed'], [left_footed, right_footed], color=['blue', 'green'])
plt.xlabel('Favorable Foot')
plt.ylabel('Number of Players')

plt.show()
```

```python
plt.figure(figsize= (10, 8))

sns.histplot(df['Overall rating'], kde = True, color= 'blue', bins= 30)
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

stats.probplot(df['Overall rating'], dist="norm", plot=plt)
plt.title("Q-Q Plot of Overall Rating")
plt.ylabel("Quantiles of Overall Rating")

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(x= 'Overall rating', y= 'Potential', data= df, color= 'lightgreen')
plt.xlabel('Overall Rating')
plt.ylabel('Potential')

plt.show()
```

```python
df.info()
```

```python
age_groups = pd.cut(df['Age'], bins=[15, 20, 25, 30, 35, 40], labels=['15-20', '21-25', '26-30', '31-35', '36-40'])
avg_rating_by_age = df.groupby(age_groups)['Overall rating'].mean()

plt.figure(figsize=(10, 6))
avg_rating_by_age.plot(kind='bar', color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Average Overall Rating')
plt.show()
```

```python
sns.pairplot(df[['Finishing', 'Dribbling', 'Short passing', 'Overall rating']], diag_kind='kde', palette='husl')
plt.suptitle('Pair Plot of Attacking Skills', y=1.02)
plt.show()
```

```python
df.info()
```

```python
plt.figure(figsize=(14, 10))

correlation_matrix = df[['Overall rating', 'Potential', 'Finishing', 'Heading accuracy', 
                         'Short passing', 'Dribbling', 'Acceleration', 'Sprint speed', 
                         'Shot power', 'Stamina', 'Strength']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')

plt.show()
```

```python
avg_finishing_by_age = df.groupby(age_groups)['Finishing'].mean()
avg_strength_by_age = df.groupby(age_groups)['Strength'].mean()

x = np.arange(len(avg_finishing_by_age))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, avg_finishing_by_age, width, label='Finishing', color='purple')
plt.bar(x + width/2, avg_strength_by_age, width, label='Strength', color='orange')

plt.xlabel('Age Group')
plt.ylabel('Average Skill Level')
plt.title('Average Finishing vs Strength by Age Group')
plt.xticks(x, avg_finishing_by_age.index)
plt.legend()

plt.show()
```

```python
contingency_table = pd.crosstab(age_groups, [df['foot_Left'], df['foot_Right']], rownames=['Age Group'], colnames=['Left Footed', 'Right Footed'])

print(contingency_table)
```
